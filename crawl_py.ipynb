```python
import uuid
import hashlib
import random
import re

def replace_func(match):
    """
    根据条件生成随机16进制字符
    :param match:
    :return:
    """
    if match.group(0) == 'x':
        return format(random.randint(0, 15), 'x')
    elif match.group(0) == 'y':
        return format(random.randint(8, 15), 'x')
def get_uuid():
    input_string = "xxxxxxxx-xxxx-4xxx-yxxx-xxxxxxxxxxxx"
    return re.sub('[xy]', replace_func, input_string)
def calculate_md5(input_string):
    return hashlib.md5(input_string.encode()).hexdigest()
def get_request_param(t):
    # _nonce_pre = str(uuid.uuid4())
    _nonce_pre = get_uuid()
    n = calculate_md5(_nonce_pre + "ad09f5f9c8b2")[:3]
    _nonce = "".join(_nonce_pre.split("-")) + n
    t["_nonce"] = _nonce
    t["_timestamp"] = str(int(time.time()))
    t['_traceId'] = "2024032411022914512626464041ee99"
    # 对字典的键进行排序并拼接成字符串
    s = ''.join(str(t[key]) + key for key in sorted(t.keys()))
    t["_sign"] = calculate_md5(s).upper()[:20]
    return t
r = {
    "catalogId": "201887",
    "pkgId": "5111",
    "pkgIdSign": "38d2e6",
    "_deviceid": "yd3962e94xgipbac5x5aa"
}
t = get_request_param(r)
t
```

```python
origin_nums = answer_datas['originNumInfos']
answers = answer_datas['paperCutItemDTOS']
from reportlab.lib.pagesizes import letter
from reportlab.platypus import SimpleDocTemplate, Paragraph, Image
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.lib.pagesizes import A4
from reportlab.lib.styles import ParagraphStyle
from reportlab.lib import colors

from reportlab.pdfbase.ttfonts import TTFont
from reportlab.pdfbase import pdfmetrics
from PIL import Image as Img
import os
pdfmetrics.registerFont(TTFont("SIMLI", r"C:\Windows\Fonts\SIMLI.TTF"))

import validators
# 加载中文字体
# fonts.addMapping('SimHei', 0, 0, 'SimHei')  # 映射正常样式
# fonts.addMapping('SimHei', 0, 1, 'SimHei-Bold')  # 映射粗体样式
# fonts.addMapping('SimHei', 1, 0, 'SimHei-Oblique')  # 映射斜体样式

def down_img(image_url, filename):
    if os.path.exists(filename):
        print(filename, '已存在')
        return
    try:
        with open(filename, "wb") as f:
            response = requests.get(image_url)
            if response.status_code != 200:
                raise Exception("图片下载失败")
            f.write(response.content)
    except Exception as ex:
        os.remove(filename)
        raise ex
    time.sleep(1)

# 创建一个 SimpleDocTemplate 对象
pdf = SimpleDocTemplate(fr"E:\PythonProject\PycharmProjects\pythonProject\test\{paper_name}.pdf", pagesize=letter)
# 定义样式
styles = getSampleStyleSheet()
# style = styles["Normal"]
style_bold_red = ParagraphStyle('BoldRed', parent=styles['Normal'])
style_bold_red.textColor = colors.red
style_bold_red.fontName = 'Helvetica-Bold'  # 使用加粗字体
style_bold_red.fontName = "SIMLI"
# 添加内容
content = []
x_offset = 20
y_offset = 20
for num, answer in zip(origin_nums, answers):
    answer = answer['answerInfo']
    # content.append(Paragraph(num, style_bold_red))
    # 保存题目
    filename = "tmp/" + answer['itemUrl'].split('/')[-1]
    down_img(answer['itemUrl'], filename)
    img = Img.open(filename)
    width, height = img.size
    # if width > 420:
    #     width = 420
    # if height > 300:
    #     height = 300
    content.append(Image(filename, width=width * 0.4, height=height * 0.4))

    print(num, answer['answer'])
    if validators.url(answer['answer']):
        urls = answer['answer'].split(',')
        for url in urls:
            filename = "tmp/" + url.split('/')[-1]
            down_img(url, filename)
            content.append(Image(filename))
        pass
    else:
        content.append(Paragraph(answer['answer'], style_bold_red))

# 将内容添加到 PDF 文档
pdf.build(content)
print('下载完成！')
```
